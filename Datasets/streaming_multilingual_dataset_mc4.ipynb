##MC4 data streaming: https://huggingface.co/datasets/allenai/c4

```python
from datasets import load_dataset, interleave_datasets

proportions = {
    'en': 0.5,
    'es': 0.3,
    'fr': 0.2
}

iters = []

for lang, _ in proportions.items():
    data = load_dataset("allenai/c4", lang, split='train', streaming=True)
    # Remove the column causing the type mismatch
    data = data.remove_columns("timestamp")
    iters.append(data) # list of each language's iterables

final_data = interleave_datasets(
    iters,
    probabilities=list(proportions.values()),
    seed=42
)
```

```python
sample = final_data.take(5)
sample = sample.map(lambda x: {'len': len(x['text'])})

next(iter(sample))
```

```python
for row in sample:
    print(row)
```

##Oscar dataset streaming: https://huggingface.co/datasets/oscar-corpus/community-oscar

###The task is extremely similar!

```python
from datasets import load_dataset, interleave_datasets

proportions = {
    'en': 0.5,
    'es': 0.3,
    'fr': 0.2
}

iters = []

for lang, _ in proportions.items():
    data = load_dataset("oscar-corpus/OSCAR-2201", lang, split='train', streaming=True)
    data.select_columns(['text']) # keep only the text column for each language
    iters.append(data) # list of each language's iterables

final_data = interleave_datasets(
    iters,
    probabilities=list(proportions.values()),
    seed=42
)
```

```python
sample = final_data.take(5)

for row in sample:
    print(row)
```
